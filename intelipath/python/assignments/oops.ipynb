Q1. Write a Python program to demonstrate multiple inheritance.
1. Employee class has 3 data members EmployeeID, Gender (String), Salary and
PerformanceRating(Out of 5) of type int. It has a get() function to get these details from
the user.
2. JoiningDetail class has a data member DateOfJoining of type Date and a function
getDoJ to get the Date of joining of employees.
3. Information Class uses the marks from Employee class and the DateOfJoining date
from the JoiningDetail class to calculate the top 3 Employees based on their Ratings
and then Display, using readData, all the details on these employees in Ascending
order of their Date Of Joining.

```python
from datetime import date, datetime
import csv as csv


class Employee:
       

    def get_from_csv(self, employeeId : str,  gender : str, salary : int, performance_rating: int, date_string:str):
        self.employeId = employeeId
        self.gender = gender
        self.salary = salary
        self.performance_rating = performance_rating
        self.joining_detail = JoiningDetail()
        self.joining_detail.set_DoJ_from_csv(date_string)
        
               
    def get(self):
        self.employeId = input('Enter the employee id : ')
        self.gender = input('Enter your gender: ')
        self.salary = int(input('Enter your salary: '))
        self.performance_rating = int(input('Enter your performance rating: '))
        self.joining_detail = JoiningDetail()
        self.joining_detail.getDoJ()


    def read_data(self):
        print('employee_details : ', self.__dict__)
        self.joining_detail.read_data()
    

class JoiningDetail:
     
    # Method to take input from the user

    def getDoJ(self):

        date_entry = input('Enter a date in YYYY-MM-DD format: ')
        year, month, day = map(int, date_entry.split('-'))
        self.joining_date = date(year=year, month=month, day=day)
    
    
    def set_DoJ_from_csv(self, date_str:str):
        year, month, day = map(int, date_str.split('-'))
        self.joining_date = date(year=year, month=month, day=day)

    # Method to print the values of class properties
    def read_data(self):
        print('Joining Details : ', self.__dict__)



class Information:

    # Create  a list to store the employess

    __employes = []
    

    def __init__(self) -> None:
        pass
    

    # Method to add an employee in the list

    def add_employee(self, emp:Employee):
        self.__employes.append(emp)
        

    def print_employes(self)->None:
        for emp in self.__employes:
            print(emp.__dict__)
    
    def get_sorted_list_by_rating(self, descending:bool):
        return sorted(self.__employes, key= lambda x:x.performance_rating, reverse=descending)
    

def read_from_csv()->Information:
    employee_information  = Information()
    with open('data/employee_details.csv', newline='') as employee_csv:
        employee_reader = csv.reader(employee_csv, delimiter=',', quotechar='|')
        for row in employee_reader:
            emp = Employee()
            emp.get_from_csv(row[0], row[1], row[2], row[3], row[4])
            employee_information.add_employee(emp)
    return employee_information


def read_from_user()->Information:
    employee_information  = Information()
    no_of_employes = int(input('Enter the number of employee you want : '))
    for i in range(0, no_of_employes):
        emp = Employee()
        emp.get()
        employee_information.add_employee(emp)
    return employee_information
         

def get_employes_by_rating()-> None:
    information = read_from_user()
    sorted_list = information.get_sorted_list_by_rating(False)
    for emp in sorted_list:
        emp.read_data()


get_employes_by_rating()
```

Q4. Create a simple Card game in which there are 8 cards which are randomly chosen from a
deck. The first card is shown face up. The game asks the player to predict whether the next card
in the selection will have a faster or lower value than the currently showing card.

For example, say the card that’s shown is a 3. The player chooses “higher,” and the next card is
shown. If that card has a higher value, the player is correct. In this example, if the player had
chosen “lower,” they would have been incorrect. If the player guesses correctly, they get 20
points. If they choose incorrectly, they lose 15 points. If the next card to be turned over has the
same value as the previous card, the player is incorrect.


```python
import random as rand

class CardGame:
    __no_of_cards = 8
    
    def __get_card(self)->int:
        return rand.randint(1, self.__no_of_cards)
    
    def __get_user_input(self)->str:
        user_input = str(input('Please provide your prediction (higher/lower) : '))
        return user_input
    
    def play_game(self)->int:
        user_input = self.__get_user_input()
        first_card_value = self.__get_card()
        next_card_value = self.__get_card()
        score = -1
        if ((user_input.lower() == 'higher' and next_card_value > first_card_value) or
            (user_input.lower() == 'lower' and next_card_value < first_card_value)):
            score = 20
        else:
            score = -15
        print('First card value : ', first_card_value)
        print('Second card value : ', next_card_value)
        print('Your choice : ', user_input)
        print('Score : ', score)
        
card_game = CardGame()
card_game.play_game()
    
```

Q6. Show a basic implementation of abstraction in python using the abstract classes.
1. Create an abstract class in python.
2. Implement abstraction with the other classes and base class as abstract class.


```python
from abc import ABC, abstractmethod

class Car(ABC):
    
    @abstractmethod
    def display_car(self):
        pass

class Altroz(Car):
    def __init__(self, color='Gold', model='XM', fuel='Petrol', chesis_no=''):
        self.color = color
        self.model = model
        self.fuel_type =fuel
        self.chesis_no = chesis_no
    
    def display_car(self):
        print(self.__dict__)
        
class Punch(Car):
    def __init__(self, color='Gold', model='XM', fuel='Petrol', chesis_no=''):
        self.color = color
        self.model = model
        self.fuel_type =fuel
        self.chesis_no = chesis_no
    
    def display_car(self):
        print(self.__dict__)
        
altroz = Altroz('Red','XM+','Disel', 'khkhdauerhhsdhs')
altroz.display_car()
punch = Punch('Green',fuel='Petrol',chesis_no='asagsjajsausua')
punch.display_car()
```

Q8. Given a list of 50 natural numbers from 1-50. Create a function that will take every element
from the list and return the square of each element. Use the python map and filter methods to
implement the function on the given list.

```python
def square(source:list)->list:
    return list(map(lambda x: x**2, source))
    
my_list = list(range(1,51))
print(my_list)
my_square_list = square(my_list)
print(my_square_list)
```
